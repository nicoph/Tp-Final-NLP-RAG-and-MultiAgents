```python
%%writefile scrap2.py

import scrapy

class MySpider(scrapy.Spider):
    name = 'p_spider'
    start_urls = ['https://www.rosarioencartel.com.ar/']  # URL de la página que deseas scrapear

    def parse(self, response):

        paragraphs_semana = response.css('.destacados_semana')
        paragraphs_proximos = response.css('.destacados_proximamente')


        with open('paragraphs_semana.html', 'w', encoding='utf-8') as f:

            for paragraph in paragraphs_semana:
                f.write(paragraph.get() + '\n')  # Escribir el texto del párrafo seguido de un salto de línea
        with open('paragraphs_proximos.html', 'w', encoding='utf-8') as f:

              for paragraph in paragraphs_proximos:
                  f.write(paragraph.get() + '\n')  # Escribir el texto del párrafo seguido de un salto de línea

```

```python
import csv
from bs4 import BeautifulSoup

with open('paragraphs_proximos.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')

resultados = []

for a_tag in soup.find_all('a'):
    espectaculo = a_tag.find(class_='title').text.strip()
    link = a_tag.get('href')
    titulo = a_tag.find(class_='title').text.strip()
    fecha = a_tag.find(class_='fecha').text.strip()
    lugar = a_tag.find(class_='place').text.strip()
    tipo_espectaculo = a_tag.find(class_='tag').text.strip()

    # Agregar los resultados a la lista para pasar a csv
    resultados.append([espectaculo, link, titulo, fecha, lugar, tipo_espectaculo])


# Escribir los resultados en un archivo CSV
with open('data/proximos.csv', 'w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['espectaculo','link del espectaculo', 'titulo del espectaculo', 'fecha del espectaculo', 'lugar del espectaculo', 'tipo de espectáculo'])
    writer.writerows(resultados)
```

#**Cargar documentos**

```python
documentos=[]
for file in os.listdir("data"):
  if file.endswith(".pdf"):
    pdf_path="./data/"+file
    loader=PyPDFLoader(pdf_path)
    documentos.extend(loader.load())
  elif file.endswith('.docx') or file.endswith('.doc'):
    doc_path="./data/"+file
    loader=Docx2txtLoader(doc_path)
    documentos.extend(loader.load())
  elif file.endswith('.txt'):
    text_path="./data/"+file
    loader=TextLoader(text_path)
    documentos.extend(loader.load())
  elif file.endswith('.csv'):
    text_path="./data/"+file
    loader = CSVLoader("./data/"+file, encoding="utf-8")
    documentos.extend(loader.load())




```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, TrainingArguments, GenerationConfig
import torch
#mas pequeño
#model_name = "clibrain/Llama-2-7b-ft-instruct-es-gptq-4bit"
model_name='clibrain/Llama-2-7b-ft-instruct-es'


#Tokenizer

tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)

def create_and_prepare_model():
        compute_dtype = getattr(torch, "float16")
        bnb_config = BitsAndBytesConfig(
            load_in_4bit=True,
            bnb_4bit_quant_type="nf4",
            bnb_4bit_compute_dtype=compute_dtype,
            bnb_4bit_use_double_quant=True,
        )
        model = AutoModelForCausalLM.from_pretrained(
            model_name, quantization_config=bnb_config, device_map={"": 0}
        )
        return model
model = create_and_prepare_model()
```

#**Hugging Face Pipeline**

```python
pipe=pipeline("text-generation",
              model=model,
              tokenizer=tokenizer,

              torch_dtype=torch.bfloat16,
              device_map='auto',
              max_new_tokens=512,
              min_new_tokens=-1,

              num_return_sequences=1,

              )
```
